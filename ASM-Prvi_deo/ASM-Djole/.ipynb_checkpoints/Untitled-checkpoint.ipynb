```python
import pandas as pd

dataPath18 = "data/Lichess/atp_matches_2018.csv"
dataPath19 = "data/Lichess/atp_matches_2019.csv"
dataPath20 = "data/Lichess/atp_matches_2020.csv"

data18 = pd.read_csv(dataPath18)
data19 = pd.read_csv(dataPath19)
data20 = pd.read_csv(dataPath20)

print(data18)
print(data19)
print(data20)
```

```python
import numpy as np

#***Koliko je puta pobedio odredjeni igrac

pobednik = 104925
odredjeni_pobedio_maska = data18['winner_id'] == pobednik 
data_black = data18[odredjeni_pobedio_maska]
labele_kolona = ['winner_id','count']
data_selektovano = data_black[labele_kolona].groupby('winner_id')
print(data_selektovano.agg(np.size))

#***Koliko je puta izgubio odredjeni igrac

gubitnik = 'Novak Djokovic'
odredjeni_izgubio_maska = data18['loser_name'] == gubitnik 
data_black = data18[odredjeni_izgubio_maska]
labele_kolona = ['loser_name','count']
data_selektovano = data_black[labele_kolona].groupby('loser_name')
print(data_selektovano.agg(np.size))

#***Sortirani parovi po broju meseva

data18["count"] = 0
kolone = ['winner_name', 'count']
grupisani = data18[kolone].groupby(['winner_name']).count()
grupisani.sort_values('count', ascending=False)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

#2018
data_reduced = data18[['winner_name', 'loser_name']]
G18 = nx.Graph()

list_winner = set(data18['winner_name'].unique())
list_loser = set(data18['loser_name'].unique())

skup_id = list_winner.union(list_loser)

G18.add_nodes_from(skup_id)

for _, winner_id, loser_id in data_reduced.itertuples():
    if (winner_id, loser_id) in G18.edges:
        G18.edges[winner_id, loser_id]['weight'] += 1
    else:
        G18.add_edge(winner_id, loser_id, weight=1)
        
#print(G.edges.data("weight"))
```

```python
#2019
data_reduced = data19[['winner_name', 'loser_name']]
G19 = nx.Graph()

list_winner = set(data19['winner_name'].unique())
list_loser = set(data19['loser_name'].unique())

skup_id = list_winner.union(list_loser)

G19.add_nodes_from(skup_id)

for _, winner_id, loser_id in data_reduced.itertuples():
    if (winner_id, loser_id) in G19.edges:
        G19.edges[winner_id, loser_id]['weight'] += 1
    else:
        G19.add_edge(winner_id, loser_id, weight=1)
```

```python
#2020
data_reduced = data20[['winner_name', 'loser_name']]
G20 = nx.Graph()

list_winner = set(data20['winner_name'].unique())
list_loser = set(data20['loser_name'].unique())

skup_id = list_winner.union(list_loser)

G20.add_nodes_from(skup_id)

for _, winner_id, loser_id in data_reduced.itertuples():
    if (winner_id, loser_id) in G20.edges:
        G20.edges[winner_id, loser_id]['weight'] += 1
    else:
        G20.add_edge(winner_id, loser_id, weight=1)
```

```python
output_path = "models/undirected_count_aggregation_2020.gml"

nx.write_gml(G20, output_path)
```

```python
data_reduced18 = data18[['winner_name', 'loser_name']]
data_reduced19 = data19[['winner_name', 'loser_name']]
data_reduced20 = data20[['winner_name', 'loser_name']]

G = nx.Graph()

list_winner18 = set(data18['winner_name'].unique())
list_loser18 = set(data18['loser_name'].unique())

list_winner19 = set(data19['winner_name'].unique())
list_loser19 = set(data19['loser_name'].unique())

list_winner20 = set(data20['winner_name'].unique())
list_loser20 = set(data20['loser_name'].unique())

skup_id18 = list_winner18.union(list_loser18)
skup_id19 = list_winner19.union(list_loser19)
skup_id20 = list_winner20.union(list_loser20)

skup_id = skup_id18.union(skup_id19).union(skup_id20)

G.add_nodes_from(skup_id)

data_reduced18 = data18[['winner_name', 'loser_name']]
data_reduced19 = data19[['winner_name', 'loser_name']]
data_reduced20 = data20[['winner_name', 'loser_name']]

for _, winner_id, loser_id in data_reduced18.itertuples():
    if (winner_id, loser_id) in G.edges:
        G.edges[winner_id, loser_id]['weight'] += 1
    else:
        G.add_edge(winner_id, loser_id, weight=1)
        
for _, winner_id, loser_id in data_reduced19.itertuples():
    if (winner_id, loser_id) in G.edges:
        G.edges[winner_id, loser_id]['weight'] += 1
    else:
        G.add_edge(winner_id, loser_id, weight=1)

for _, winner_id, loser_id in data_reduced20.itertuples():
    if (winner_id, loser_id) in G.edges:
        G.edges[winner_id, loser_id]['weight'] += 1
    else:
        G.add_edge(winner_id, loser_id, weight=1)
        
print(G.edges.data("weight"))
```

```python
output_path = "models/undirected_count_aggregation.gml"

nx.write_gml(G, output_path)
```

```python
# 1.Pitanje

degrees = G.degree()
sumica = sum(deg for node, deg in degrees)
print(sumica/len(G))


degrees18 = [(node, degree) for (node, degree) in G18.degree()]



degrees19 = [(node, degree) for (node, degree) in G19.degree()]

degrees20 = [(node, degree) for (node, degree) in G20.degree()]

degrees_year_avg = []

for player_name in skup_id:
    degrees_year_avg.append((player_name, 
                             (next((degree for (node, degree) in degrees18 if node == player_name), 0) +
                                 next((degree for (node, degree) in degrees19 if node == player_name), 0) + 
                                    next((degree for (node, degree) in degrees20 if node == player_name), 0))  / 3))
    
    
print(degrees_year_avg)

#print(degrees)
```
